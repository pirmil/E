```python
# Sample DataFrames
X_train_data = {
    'Date (UTC)': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'Value': [10, 20, 30]
}

production_data = {
    'Date': ['2024-01-01', '2024-01-03'],
    'Production': [100, 200]
}

X_train = pd.DataFrame(X_train_data)
production_mensuelle_energie_eolienne = pd.DataFrame(production_data)

# Initial X_train
print("Initial X_train:")
display(X_train)

# Initial production_mensuelle_energie_eolienne
print("\nInitial production_mensuelle_energie_eolienne:")
display(production_mensuelle_energie_eolienne)

# Merge and display the result
print("\nMerged on right(prod_mens decides the index):")
display(X_train.merge(production_mensuelle_energie_eolienne, left_on='Date (UTC)', right_on='Date', how='right'))

print("\nMerged on left(X_train decides the index):")
display(X_train.merge(production_mensuelle_energie_eolienne, left_on='Date (UTC)', right_on='Date', how='left'))
```

```python
import pandas as pd

# Original DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data, index=pd.Series(range(1, len(data) + 1), name='Date'))

display(df)

# Melt the DataFrame to reshape it
melted_df = df.reset_index().melt(id_vars='Date', var_name='Variable', value_name='Value')

display(melted_df)

# Combine Date and Variable columns to create the new index
melted_df['Date_and_hour'] = melted_df['Date'].astype(str) + melted_df['Variable']

# Sort the DataFrame based on the new index
melted_df.sort_values(by='Date_and_hour', inplace=True)

# Set the new index
melted_df.set_index('Date_and_hour', inplace=True)

# Drop unnecessary columns
melted_df.drop(columns=['Date', 'Variable'], inplace=True)

display(melted_df)
```
